```python
import pandas as pd
```

```python
df = pd.read_html('https://kworb.net/spotify/country/us_daily_totals.html')[0]
```

```python
clean_df = pd.DataFrame()
clean_df[['Artist','Title']] = df['Artist and Title'].str.split(' - ',expand = True, n = 1)
clean_df['Streams'] = (df['Total']/1000000).round(decimals = 0).astype(int)
```

```python
df.head()
```

```python
clean_df['h-index'] = clean_df.groupby('Artist')['Streams'].transform(lambda x: ( x >= x.rank(ascending=False, method='first')).sum())
```

```python
clean_df.head()
```

```python
artist_stats = clean_df.groupby('Artist').agg({'Streams':'sum', 'Title':'nunique', 'h-index':'mean'})
artist_stats.rename(columns = {'Title':'Song_Count'}, inplace = True)
```

```python
artist_stats['h-index_avg'] = (100*artist_stats['h-index'] / artist_stats['Song_Count']).round(decimals = 2)
```

```python
artist_stats.sort_values('h-index', ascending = False, inplace = True)
```
